```python
import numpy as np
import pandas as pd

file_path = "traffic_accident_2020.csv"
df = pd.read_csv(file_path, encoding='CP949')
df.head()
```

```python
df.info()
df.describe()
```

```python
df.columns.values
```

```python
df.index.values
```

```python
# 불필요한 컬럼 삭제 /// axis=1 : 컬럼삭제   &   inplace=True : 원본데이터 수정
# 상관계수를 판단하고, 그에 따라 불필요/필요 여부를 판단하면 더 정확하다.
df.drop(['발생년','사고유형','가해자법규위반','발생위치X_UTMK','발생위치Y_UTMK'],axis=1,inplace=True)
```

```python
df.describe()
```

```python
# 컬럼명 변경
df.rename(columns={'발생년월일시':'월일시','사망자수':'사망','사상자수':'사상','중상자수':'중상','경상자수':'경상','부상신고자수':'부상',
                  '사고유형_대분류':'유형A','사고유형_중분류':'유형B','도로형태_대분류':'도로A','도로형태':'도로B',
                  '피해자_당사자종별':'피해','가해자_당사자종별':'가해'},inplace=True)
df.columns.values
```

```python
# 컬럼 새로 추가
df['월']=df['월일시'].astype(str).str[4:6]
df['일']=df['월일시'].astype(str).str[6:8]
df['시']=df['월일시'].astype(str).str[8:]

# 월일시 > 월/일/시 로 쪼개면서 [월일시] 컬럼이 필요가 없어졌으므로 삭제한다.
df.drop(['월일시'],axis=1,inplace=True)
df.head()
```

```python
# 결측치 확인 : isnull()
missing_df = df.isnull()
missing_type = df[['피해','가해']].isnull()

missing_type.describe() # False로 null값 없음을 확인할 수 있다.
```

```python
# 중복없이 unique한 값출력 : unique()
# 고유값별로 데이터 개수 계산 : value_counts()
# 가장 개수가 많은 데이터 고유값 출력 : idxmax()

df['피해'].unique()
df['피해'].value_counts()
df['피해'].value_counts().idxmax()
```
