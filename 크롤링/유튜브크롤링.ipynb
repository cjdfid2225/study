유튜브는 스크롤을 내리지 않으면 모든 영상들이 page source에 보이지 않기 때문에 스크롤을 내리는 작업을 선행해준다.

20회 정도만 스크롤을 내린 후 작업 진행.

```python
body = browser.find_element_by_tag_name('body') #스크롤하기 위해 소스 추출

num_of_pagedowns = 20
while num_of_pagedowns:
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
    num_of_pagedowns -= 1
```

selenium은 굉장히 오래걸리기 때문에 selenium으로 필요한 부분까지만 진행하고 나머지는 beautiful soup, request로 진행.

```python
html0 = browser.page_source
html = BeautifulSoup(html0, 'html.parser')
```

```python
video_list0 =html.find_all('ytd-grid-video-renderer',{'class':'style-scope ytd-grid-renderer'})
video_list1 = html.find('div',{'id':'items','class':'style-scope ytd-grid-renderer'})
len(video_list1.find_all('ytd-grid-video-renderer',{'class':'style-scope ytd-grid-renderer'}))
```
