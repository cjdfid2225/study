# where()

for loop & if 보다 속도가 빠르다. 

조건에 맞는 값의 인덱스를 뽑아주거나, 조건에 맞는 값을 다른 값으로 대체하는 함수.

```python
arr1 = np.array([10,30,50,20,80,40,50])
print(arr1.min(), np.min(arr1))
print(arr1.max(), np.max(arr1))
print(arr1.argmin(), np.argmin(arr1), arr1.argmax(), np.argmax(arr1)) # argmin(), argmax() : output이 인덱스
result = np.where(arr1>50)
print(result, type(result))
print(result[0])
print(result[0][0])

result = arr1[np.where(arr1>50)]
print(result)

print(np.where(arr1>50, 15, arr1)) # 이러면 arr1에 결과반영x == 15로 바뀌지x
result = np.where(arr1>50, 15, arr1)
print(result, arr1)
```

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','sex','class','fare','survived']]
display(df.head())
```

```python
# class열을 기준으로 분할
grouped = df.groupby(['class'])
print(grouped)

for key, group in grouped:
    print('*key: ', key)
    print('*number: ', len(group))
    display(group.head())
```

```python
# 연산메소드 적용
average = grouped.mean()
min_ = grouped.min()
display(average)
display(min_)
print()

# 개별 그룹 선택하기
group3 = grouped.get_group('Third')
display(group3.head())
```

```python
# class열, sex열을 기준으로 분할
grouped_two = df.groupby(['class','sex'])

# grouped_two 그룹 객체를 iteration으로 출력
for key, group in grouped_two:
    print('* key: ',key)
    print('* number: ',len(group))
    display(group.head())
    print()
    
# grouped_two 그룹 객체에 연산 메소드 적용
average_two = grouped_two.mean()
display(average_two)
print()
print(type(average_two))

# grouped_two 그룹 객체에서 개별 그룹 선택하기
group3f = grouped_two.get_group(('Third','female'))
display(group3f.head())
```

```python
average_age = grouped_two.age.mean()
display(average_age)
```

```python
grouped = df.groupby(['class'])

# 각 그룹에 대한 모든 열의 표준편차를 집계하여 데이터프레임으로 반환
std_all = grouped.std()
print(std_all)
print()
print(type(std_all))
print()

# 각 그룹에 대한 fare 열의 표준편차를 집계하여 시리즈로 반환
std_fare = grouped.fare.std()
print(std_fare)
print()
print(type(std_fare))
print()
```

```python
# 그룹 객체에 agg() 메소드 적용_ 사용자 정의 함수를 인수로 전달
def min_max(x):
    return x.max() - x.min()

# 각 그룹의 최대값과 최소값의 차이를 계산하여 그룹별로 집계
agg_minmax = grouped.agg(min_max)
print(agg_minmax.head())
print()

# 여러 함수를 각 열에 동일하게 적용하여 집계
agg_all = grouped.agg(['min','max'])
print(agg_all.head())
print()

# 각 열마다 다른 함수를 적용하여 집계
agg_sep = grouped.agg({'fare':['min','max'],'age':'mean'})
print(agg_sep.head())
```

```python
# 그룹 객체에 filter 함수 적용
# class 열을 기준으로 분할
grouped = df.groupby(['class'])

# 데이터 개수가 200개 이상인 그룹만을 필터링하여 데이터프레임으로 반환
grouped_filter = grouped.filter(lambda x: len(x) >= 200)
display(grouped_filter.head())
print()
print(type(grouped_filter))

# age열의 평균이 30보다 작은 그룹만을 필터링하여 데이터프레임으로 반환
age_filter = grouped.filter(lambda x: x.age.mean() < 30)
display(age_filter.tail())
print()
print(type(age_filter))

# 30 이하의 평균 및 30 이하의 group을 성별로 다시 재그룹
print(age_filter.mean())
age_filter_sex = age_filter.groupby(['sex'])
display(age_filter_sex.mean())
```

```python
# 피벗 예
import pandas as pd
import seaborn as sns
from IPython.display import display

# titanic 데이터셋에서 age, sex 등 5개 열을 선택하여 데이터프레임 만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','sex','class','fare','survived']]
display(df.head())
print()

# 행, 열, 값, 집계에 사용할 열을 1개씩 지정 - 평균 집계
pdf1 = pd.pivot_table(df,
                     index='class',
                     values='survived',
                     aggfunc='mean')
display(pdf1.head())
```

```python
pdf1 = pd.pivot_table(df,
                     index='class',
                     columns='sex',
                     values='age',
                     aggfunc='mean')
display(pdf1.head())
print()

# 값에 적용하는 집계 함수를 2개 이상 지정 가능 - 생존율, 생존자 수 집계
pdf2 = pd.pivot_table(df,
                     index='class',
                     columns='sex',
                     values='survived',
                     aggfunc=['mean','sum'])
display(pdf2.head())
```

```python
# 행, 열 값에 사용할 열을 2개 이상 지정 가능- 평균나이, 최대요금 집계
pdf3 = pd.pivot_table(df,
                     index=['class','sex'],
                     columns='survived',
                     values=['age','fare'],
                     aggfunc=['mean','max'])

# IPython Console 디스플레이 옵션 설정
pd.set_option('display.max_columns',10) #출력할 열의 개수 한도
display(pdf3.head())
print()

# 행, 열 구조 살펴보기
print(pdf3.index)
print(pdf3.columns)
```

```python
print(pdf3.xs('First'))
print(pdf3.xs(('First','female')))
print(pdf3.xs('male',level='sex'))
print(pdf3.xs(('Second','male'),level=[0,'sex']))
```

```python
# Titanic에서 age, fare 컬럼에서 결측치를 확인하는 코드 작성
df = titanic.loc[:,['age','fare']]

def missing_value(data):
    return data.isnull()

missing_result = df.apply(missing_value, axis=0)
display(missing_result)
print(missing_result.sum())
```
