```python
import urllib.request
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = 'https://www.kamis.or.kr/customer/price/wholesale/item.do?action=priceinfo&regday=2020-12-31&itemcategorycode=600&itemcode=644&kindcode=&productrankcode=&convert_kg_yn=N'
url_source = urllib.request.urlopen(url)
soup = BeautifulSoup(url_source,'html.parser',from_encoding='utf-8')
tables = soup.select('.wtable3') #class='wtable3'인 것만 가져오니까 클래스는 .클래스명 으로 가져오면 된다

# 보기 좋게 df형식으로 바꾸는 과정~
table = tables[0]
table_html = str(table)
table_df_list = pd.read_html(table_html)
table_df = table_df_list[0]
```

## 최종판

```python
# 굴 > itemcategorycode=644
# 전복 > itemcategorycode=653

url_644 = 'https://www.kamis.or.kr/customer/price/wholesale/item.do?action=priceinfo&regday=' + str(url_date) + '&itemcategorycode=600&itemcode=644&kindcode=&productrankcode=&convert_kg_yn=N'
for m in range(365):
    
    day = timedelta(days=1)
    url_date = date(2019,12,31) + day
    url_date = url_date.isoformat()
    url_source = urllib.request.urlopen(url_644)
    soup = BeautifulSoup(url_source,'html.parser',from_encoding='utf-8')
    tables = soup.select('.wtable3') 
    table = tables[0]
    table_html = str(table)
    table_df_list = pd.read_html(table_html)
    table_df = table_df_list[0]
    
    table_df.to_csv('price_굴_'+str(m)+'.csv')
    
url_653 = 'https://www.kamis.or.kr/customer/price/wholesale/item.do?action=priceinfo&regday=' + str(url_date) + '&itemcategorycode=600&itemcode=653&kindcode=&productrankcode=&convert_kg_yn=N'
for m in range(365):
    
    day = timedelta(days=1)
    url_date = date(2019,12,31) + day
    url_date = url_date.isoformat()
    url_source = urllib.request.urlopen(url_653)
    soup = BeautifulSoup(url_source,'html.parser',from_encoding='utf-8')
    tables = soup.select('.wtable3') 
    table = tables[0]
    table_html = str(table)
    table_df_list = pd.read_html(table_html)
    table_df = table_df_list[0]
    
    table_df.to_csv('price_전복_'+str(m)+'.csv')
```
